```python
import cv2
import random
from google.colab.patches import cv2_imshow


image = cv2.imread("//content//drive//MyDrive//rcnn1//pramid.jpg")
image = cv2.resize(image, dsize= (600,600))
cv2_imshow(image)

#ilklendirme

ss = cv2.ximgproc.segmentation.createSelectiveSearchSegmentation()  ###bu satÄ±rda hata var.
ss.setBaseImage(image)
ss.switchToSelectiveSearchQuality()


print("start")
rects = ss.process()

output = image.copy()

for (x,y,w,h) in rects[:50]:
    color = [random.randint(0,255) for j in range(0,3)]
    cv2.rectangle(output, (x,y), (x+w,y+h),color,2)

cv2_imshow(output)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
